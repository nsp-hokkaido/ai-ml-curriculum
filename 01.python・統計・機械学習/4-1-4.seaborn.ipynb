# ***Seaborn***
---
Seabornは、グラフ描画のライブラリ。<br>
内部でMatplotlibが動いているが、Matplotlibと比べて以下の違いがある。<br>
・きれいな描画でグラフが描ける。<br>
・少ないコードでグラフが描ける。<br>

### ■Seabornの使用
Seabornを利用するために、インポートする。<br>

```python
import seaborn as sns

#matplotlib.pyplotモジュールも併せてインポート
import matplotlib.pyplot as plt
```

Seabornでは、サンプル用のデータセットをDataFrameとして使用することができる。

```python
# flightsデータのデータセット
flights_data = sns.load_dataset('flights')
flights_data.head()
```

```python
# irisデータのデータセット
iris_data = sns.load_dataset('iris')
iris_data.head()
```

### ■折れ線グラフ
lineplotで、flightsデータセットから各年1月の乗客数の折れ線グラフを描画する。<br>
x軸：year（年）（日時による量的変数）<br>
y軸：passengers（乗客数）（量的変数）<br>

```python
# monthがJan（1月）のデータを取得
flights_data2 = flights_data[(flights_data['month'] == 'Jan')]

print(flights_data2)

# lineplot
sns.lineplot(x='year', y='passengers', data=flights_data2, ci=None)
plt.show()
```

### ■棒グラフ
barplotで、irisデータセットから種類ごとの花弁の長さの平均値の棒グラフを描画する。<br>
x軸：species（アヤメの種類）（質的変数）<br>
y軸：sepal_length（花弁の長さ）（量的変数）<br>

```python
# 種類ごとの花弁の長さの平均値
df = iris_data.groupby('species').mean().reset_index()

print(df)

# barplot
sns.barplot(x='species', y='sepal_length', data=df)
plt.show()
```

### ■散布図
scatterplotで、irisデータセットからアヤメの種類ごとの花弁の幅と長さの散布図を描画する。<br>
x軸 ：sepal_width（花弁の幅）（量的変数）<br>
y軸 ：sepal_length（花弁の長さ）（量的変数）<br>
凡例（hue）：species(アヤメの種類)<br>

```python
# scatterplot
sns.scatterplot(x='sepal_width', y='sepal_length', hue="species", data=iris_data)
plt.show()
```

### ■分布付きの散布図
jointplotでirisデータセットから花弁の幅と長さの分布付きの散布図を描画する。<br>
x軸 ：sepal_width（花弁の幅）（量的変数）<br>
y軸 ：sepal_length（花弁の長さ）（量的変数）<br>

```python
# jointplot
sns.jointplot(x='sepal_width', y='sepal_length', hue="species", data=iris_data)
plt.show()
```

縦軸と横軸の分布からデータが多い箇所を確認することができる。

### ■ヒストグラム
histplotで、irisデータセットからアヤメの花弁幅のヒストグラムを描画する。<br>
x軸 ：sepal_width（花弁の幅）（量的変数）<br>
y軸 ：度数<br>

```python
# histplot
sns.histplot(iris_data['sepal_width'])
plt.show()
```

### ■箱ひげ図
boxplotで、irisデータセットからアヤメの種類ごとの花弁の長さの箱ひげ図を描画する。<br>
x軸 ：アヤメの種類（質的変数）<br>
y軸 ：sepal_length（花弁の長さ）（量的変数）<br>

```python
# barplot
sns.boxplot(x=iris_data['species'], y=iris_data['sepal_length'])
plt.show()
```

### ■散布図行列
pairplotでirisデータセットから散布図行列を描画する。<br>
散布図行列は、複数のデータに対して、すべての組み合わせで作成した散布図を行列形式でまとめたもの。<br>
どのデータの組み合わせが重要な関係性を持っているか確認することができる。

```python
# pairplot
sns.pairplot(iris_data)
plt.show()
```

「凡例（hue）：species(アヤメの種類)」を指定し、色分け表示もできる。

```python
# pairplot
sns.pairplot(iris_data, hue="species")
plt.show()
```
